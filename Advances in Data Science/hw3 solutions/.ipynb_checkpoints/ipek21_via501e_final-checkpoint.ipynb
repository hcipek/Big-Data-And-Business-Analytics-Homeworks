```python
class SectionA:
    def __init__(this, autoType):
        this.type = autoType
        this.new = 0
        this.used = 0
        this.usedTotal = 0
        this.newTotal = 0
        this.maxAge = 0
        this.minAge = 9223372036854775807
        
    def addNew(this, mileage, age):
        this.new += mileage
        this.newTotal += 1
        this.checkAge(age)
        
    def addUsed(this, mileage, age):
        this.used += mileage
        this.usedTotal += 1
        this.checkAge(age)
        
    def checkAge(this, age):
        if this.maxAge < age:
            this.maxAge=age
        if this.minAge > age:
            this.minAge=age
        
    def getNew(this):
        if this.newTotal != 0:
            return this.new / this.newTotal
        else:
            return 0
    
    def getUsed(this):
        if this.usedTotal != 0:
            return this.used / this.usedTotal
        else:
            return 0
        
    def getAgeDifference(this):
        return this.maxAge - this.minAge
            
        
def problem1(path):
    import pandas as pd
    from tabulate import tabulate
    
    types = [SectionA('Large SUV'), SectionA('Mid-size'), SectionA('Minivan'), SectionA('Small'), SectionA('Small SUV')]
    
    data = pd.read_csv(path, header=0)
    
    for x in range(len(data.index)):
        df = data.iloc[x]
        genericOp(df, types)
        
    sectionAOutput = {}
    sectionBOutput = {}
    
    for x in types:
        if 'Purchased Type' in sectionAOutput:
            _l = sectionAOutput['Purchased Type']
            _l.append(x.type)
        else: 
            _l = [x.type]
        sectionAOutput['Purchased Type']=_l
        
        if 'New' in sectionAOutput:
            _l = sectionAOutput['New']
            _l.append(x.getNew())
        else: 
            _l = [x.getNew()]
        sectionAOutput['New']=_l
        
        if 'Used' in sectionAOutput:
            _l = sectionAOutput['Used']
            _l.append(x.getUsed())
        else: 
            _l = [x.getUsed()]
        sectionAOutput['Used']=_l
            
        if 'Type' in sectionBOutput:
            _l = sectionBOutput['Type']
            _l.append(x.type)
        else: 
            _l = [x.type]
        sectionBOutput['Type']=_l
            
        if 'Vehicle_Age' in sectionBOutput:
            _l = sectionBOutput['Vehicle_Age']
            _l.append(x.getAgeDifference())
        else: 
            _l = [x.getAgeDifference()]
        sectionBOutput['Vehicle_Age']=_l
        
    df = pd.DataFrame(sectionAOutput)
    print(tabulate(df, headers='keys', tablefmt='psql'))
    df = pd.DataFrame(sectionBOutput)
    print(tabulate(df, headers='keys', tablefmt='psql'))
        
def genericOp(df, types):
    exit = False
    i = 0
    while not exit and i < len(types):
        if types[i].type == df['Type']:
            if df['Purchased'] == 'Used':
                types[i].addUsed(df['Mileage'], df['Vehicle_Age'])
            else:
                types[i].addNew(df['Mileage'], df['Vehicle_Age'])
            exit = True
        i += 1
```

```python
class complaint:
    def __init__(this, cType):
        this.type = cType
        this.hours = list()
        this.total = 1
        this.closed = 0
        
    def addHours(this, hours):
        this.hours.append(hours)
        this.total += 1
        this.closed += 1
        
    def notClosed(this):
        this.total += 1

import pandas as pd
from datetime import datetime
from dateutil.parser import *

def problem3(path):
    from tabulate import tabulate
    import statistics

    
    complaints = dict()
        
    data = pd.read_csv(path, header=0)
    
    for i in range(len(data.index)):
        x = data.iloc[i]
        if x['Complaint Type'] in complaints:
            tmp = complaints[x['Complaint Type']]
        else:
            tmp = complaint(x['Complaint Type'])
        
        if x['Status']=='Closed':
            diff = difference(x['Created Date'], x['Closed Date'])
            tmp.addHours(diff/3600)
        else:
            tmp.notClosed()
        
        complaints[x['Complaint Type']]=tmp
        
    result = {'Complaint Type': [], 'mean': [], 'count': [], 'total': []}
    
    for cType, c in complaints.items():
        if c.closed != 0:
            types = result['Complaint Type']
            means = result['mean']
            counts = result['count']
            total = result['total']

            types.append(c.type)
            means.append(round(statistics.mean(c.hours), 2))
            counts.append(c.closed)
            total.append(c.total)

            result['Complaint Type'] = types
            result['mean'] = means
            result['count'] = counts
            result['total'] = total
    
    df = pd.DataFrame(result)
    df = df.sort_values(by='total', ascending=False).iloc[:10].sort_values(by='Complaint Type')
    print(tabulate(df[["Complaint Type", "mean", "count"]], headers='keys', tablefmt='psql', showindex=False))
    
def difference(create, end):
    
    cDate = parse(create)
    eDate = parse(end)
    if (cDate > eDate):
        return (eDate-cDate).total_seconds() + 60*60*12
    return (eDate-cDate).total_seconds()
```

```python
import pandas as pd

def problem4(folder):
    import os
    import pandas as pd
    from tabulate import tabulate
    
    csvDict = dict()
    
    for root, dirs, files in os.walk(folder):
        for x in files:
            nx = root + "\\" + x
            data = pd.read_csv(nx, header=0)
            _l = x[:-4].split("_")
            if _l[0] == 'olist':
                _l.pop(0)
                _l.pop(-1)
            else:
                #this data only has one csv file that dont start with 'olist', which is "product_category_name_translation"
                _l = [_l[0], _l[-1]]
            csvDict['_'.join(_l)]=data
    
    mergedDf = merger(csvDict)
    
    output = dict()
    for i, row in mergedDf.iterrows():
        if row['seller_city'] != row['customer_city']:
            string1 = '(' + row['seller_city'] + ',' + row['customer_city'] + ')'
            string2 = '(' + row['customer_city'] + ',' + row['seller_city'] + ')'
            if string1 in output:
                val = output[string1] + 1
                string = string1
            elif string2 in output:
                val = output[string2] + 1
                string = string2
            else:
                val = 1
                string = string1
            output[string] = val
    
    df = pd.DataFrame(output.items(), columns=['', 'Total_orders'])
    df = df.sort_values(by='Total_orders', ascending=False).iloc[:10]
    print(tabulate(df, headers='keys', tablefmt='psql', showindex=False))
        
def merger(csvDict):
    return pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(csvDict["order_items"],csvDict["products"]),csvDict["product_translation"]),csvDict["orders"]),csvDict["sellers"]),csvDict["customers"])
```

```python
import pandas as pd
from tabulate import tabulate

class sectionC:
    def __init__(this, cType, status):
        this.type = cType
        this.assign = 0
        this.close = 0
        this.open = 0
        this.pend = 0
        this.unassign = 0
        this.total = 0
        this.setStat(status)
            
    def setStat(this, status):
        if status == 'Assigned':
            this.assigned()
        elif status == 'Closed':
            this.closed()
        elif status == 'Open':
            this.opened()
        elif status == 'Pending':
            this.pending()
        elif status == 'Unassigned':
            this.unassigned()
            
    def assigned(this):
        this.assign += 1
        this.total += 1
        
    def closed(this):
        this.close += 1
        this.total += 1
    
    def opened(this):
        this.open += 1
        this.total += 1
    
    def pending(this):
        this.pend += 1
        this.total += 1
    
    def unassigned(this):
        this.unassign += 1
        this.total += 1

def problem5(path):
    import statistics
    
    data = pd.read_csv(path, header=0)
    
    sectionADict = dict()
    sectionBDict = dict()
    sectionCDict = dict()
    
    allCTypes = set()
    
    for i, row in data.iterrows():
        #SectionA operations
        date = parse(row['Created Date']).date()
        key = date.strftime('%B-%d-%A')
        if key in sectionADict:
            val = sectionADict[key] + 1
        else:
            val = 1
        sectionADict[key] = val
        
        #SectionB Operations
        weekday = date.strftime('%A')
        if weekday in sectionBDict:
            val = sectionBDict[weekday]
        else:
            val = dict()
        complaint = row['Complaint Type']
        allCTypes.add(complaint)
        if complaint in val:
            val2 = val[complaint] + 1
        else:
            val2 = 1
        val[complaint] = val2
        sectionBDict[weekday] = val
        
        #SectionC Operations
        stat = row['Status']
        if complaint in sectionCDict:
            c = sectionCDict[complaint]
            c.setStat(stat)
        else:
            c = sectionC(complaint, stat)
        sectionCDict[complaint] = c
    
    sectAOp(sectionADict)
    mostFreq = sectBOp(sectionBDict, allCTypes)
    sectCOp(sectionCDict, mostFreq)
    
                
def sectAOp(x):
    df = pd.DataFrame(x.items(), columns=['Date', 'Count'])
    df = df.sort_values(by='Count', ascending=False).iloc[:1]
    
    print("A)")
    print(df['Date'].iloc[0])
    
def sectBOp(x, cTypes):
    import itertools 
    
    d = {'day': []}
    for i in cTypes:
        d[i] = []
    for k, v in x.items():
        d['day'].append(k)
        for a, b in d.items():
            if a in v:
                b.append(v[a])
            elif a != 'day':
                b.append(0)
    totalDict = dict()
    for k, v in d.items():
        if k != 'day':
            totalDict[k] = sum(v)
    totalDict = dict(sorted(totalDict.items(), key=lambda item: item[1], reverse=True))
    totalDict = dict(itertools.islice(totalDict.items(), 10)) 
    totalDict = dict(sorted(totalDict.items(), key=lambda item: item[0]))
    
    tempDict = {'day': d['day']}
    
    for key in totalDict:
        tempDict[key] = d[key]
        
    df = pd.DataFrame(tempDict)
    df = df.sort_values(by='day', ascending=True)
    print("-------------------------------------------------------------------------------------")
    print("B)")
    print(df)
    
    return totalDict
            
def sectCOp(x, freq):
    tempDict = {'Complaint Type': [], 'Assigned': [], 'Closed': [], 'Open': [], 'Pending': [], 'Unassigned': []}
    
    for k, v in x.items():
        if k in freq:
            tempDict['Complaint Type'].append(k)
            tempDict['Assigned'].append(round(v.assign*100/v.total, 2))
            tempDict['Closed'].append(round(v.close*100/v.total, 2))
            tempDict['Open'].append(round(v.open*100/v.total, 2))
            tempDict['Pending'].append(round(v.pend*100/v.total, 2))
            tempDict['Unassigned'].append(round(v.unassign*100/v.total, 2))
    
    df = pd.DataFrame(tempDict)
    df = df.sort_values(by='Complaint Type', ascending=True)
    print("-------------------------------------------------------------------------------------")
    print("C)")
    print(tabulate(df, headers='keys', tablefmt='psql', showindex=False))

        
```
